```
########### BOOK USED ==> 'The Project Gutenberg eBook of The history of magic, by Éliphas'
```

```
book = sc.textFile("/FileStore/tables/ebook.txt")
```

```
#SPLITTING INTO WORDS, REMOVING STOP-WORDS AND MAKING LOWER-CASE
```

```
book = book.flatMap(lambda x:x.split(' '))
```

```
book=book.map(lambda x:x.lower())
```

```
nltk.download('averaged_perceptron_tagger')
```

```
book_postags = pos_tag(book.collect())
```

```
book_postags
```

```
book_postags = sc.parallelize(book_postags)
```

```
#EXTRACTING ONLY THE NAMED ENTITIES USING POSTAGS
```

```
N=['NN','NNP','NNS','NNPS']
```

```
bp=book_postags.filter(lambda x: x[1] in N).filter(lambda x:x[0]!='')
```

```
#bp.take(15)
```

```
named_ents = bp.map(lambda x:x[0])
```

```
#named_ents.collect()
```

```
nltk.download('stopwords')
```

```
stop_words = stopwords.words('english')
```

```
named_ents = named_ents.filter(lambda x: x not in stop_words)
```

```
named_ents=named_ents.map(lambda x:(x,1))
```

```
#named_ents.collect()
```

```
named_ents = named_ents.reduceByKey(lambda x,y: x+y)
```

```
#named_ents.collect()
```

```
sorted_ne= named_ents.sortBy(lambda x:-x[1])
```

```
#FINAL OUTPUT WORD COUNNTS SORTED IN DESCENDING ORDER
```

```
sorted_ne.collect()
```
